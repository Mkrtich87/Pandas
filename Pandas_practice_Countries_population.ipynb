```python
# Plot the pie chart for Population (2020)
df.set_index('Country (or dependency)')['Population (2020)'].plot.pie(
    autopct='%1.1f%%',  # Display percentage on slices
    figsize=(8, 8),     # Set the figure size
    startangle=90,      # Rotate the chart for better readability
    title='Population Distribution (2020)'  # Add a title
)

plt.ylabel('')  # Remove the y-axis label for a cleaner chart
plt.show()
```

```python
top_25 = df.sort_values(by='Population (2020)', ascending=False).head(25)

# Plot the top 25 countries by population as a bar chart
plt.figure(figsize=(10, 8))
plt.barh(top_25['Country (or dependency)'], top_25['Population (2020)'], color='skyblue')
plt.xlabel('Population (2020)')
plt.ylabel('Country')
plt.title('Top 25 Countries by Population (2020)')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.tight_layout()

plt.show()
```
