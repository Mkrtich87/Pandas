```python
def fillna_random(new_df):
    for col in new_df.columns:
        if new_df[col].isnull().values.any():
            # Count the number of NaN values in the column
            num_nulls = new_df[col].isnull().sum()
            # Generate random numbers to fill NaN values
            rand_values = np.random.randn(num_nulls)
            # Replace NaN values with random numbers
            new_df[col].fillna(pd.Series(rand_values), inplace=True)
    return new_df

df_filled = fillna_random(new_df)
```

Create a line plot to visualize the relationship between "Age" and "City" in the DataFrame you created earlier (from Exercise 2).
Use different colors to represent each individual.
Label the axes and add a title to the plot.
Hint: You can use the scatter function with the c parameter to specify colors.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {"Name": ["John", "Moses","Volodimir", "Denise", "Grenik"],
         "Age": [23, 12, 56, 43, 63], 
         "City": ["yerevan", "Stepanakert","Goris", "Las Vegas", "Varna"]}

df = pd.DataFrame(data)
# Create a list of colors (one for each individual)
colors = plt.cm.viridis(np.linspace(0, 1, len(df)))  # Use viridis colormap for variety

# Create the scatter plot with individual colors
plt.scatter(df["City"], df["Age"], c=colors, marker='o', edgecolors='black')

# Add legend with individual names as labels
plt.legend(df['Name'], title="Individuals", loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel("Age")
plt.ylabel("City")
plt.title("Scatter Plot Age vs City")
plt.grid(True)
plt.show

```

Plot a sine function with values ranging from 0 to 2π using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate the sine function
y = np.sin(x)

# Create the plot
plt.plot(x, y, label="sin(x)", color="b")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Function")
plt.grid(True)
plt.legend()
plt.show()
```

Create a NumPy array containing random integers between 1 and 100 with a size of (4, 5).
Convert the NumPy array to a Pandas DataFrame.
Add a new column named "Squared" to the DataFrame. This column should contain the square of each element in the original data.

```python
import numpy as np
import pandas as pd

# Create a NumPy array with random integers
arr = np.random.randint(1, 101, size=(4, 5))

# Convert the array to a DataFrame
df = pd.DataFrame(arr)

# Add a new column with squared values
df_squared = df.assign(squared=df[0] ** 2)

# Print the DataFrame with the squared values
print(df_squared)
```
